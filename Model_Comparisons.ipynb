```python
import pandas as pd
```

```python
df= pd.read_csv("train.csv")
df.head(50)
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd

train_df = df
train_df = train_df.rename(columns={'Class Index':'labels'})
train_df = train_df.rename(columns={'Description':'text'})

encoding_labels = {1:0, 2:1, 3:2, 4:3}

train_df['labels'] = train_df['labels'].map(encoding_labels)

model_args = ClassificationArgs(num_train_epochs=3, overwrite_output_dir= True)

model = ClassificationModel('bert','bert-base-cased',num_labels=12,args=model_args) 
```

```python
train_df.head(50)
```

```python
eval_df = train_df[-500:]

# Train the model
model.train_model(train_df)

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```

```python
test_df = pd.read_csv( 'haha_2021_test.csv')

test_df = pd.DataFrame({'text': test_df['text'].replace(r'\n', ' ', regex=True)})

a=[]
for i in test_df['text']:
  a.append(i)

ans = model.predict(a)
test_df['humor_mechanism']= ans[0]
del test_df['text']
test_df.head()
```
